## Funciones para la generai贸n de variables aleatorias a partir de n煤meros aleatorios

### Distribuci贸n exponencial por Rejection-Sampling

**Funci贸n inversa de la distribuci贸n exponencial**

```python
exponentialInv = lambda x,l: -np.log(1-x)/l
```

**Funci贸n para generar una variable perteneciente a la distribuci贸n exponencial**

```python
def expSim():
    x = -1
    while x == -1:
        x_s = random.uniform(0,1)
        u = random.uniform(0,1)
        if u<exponentialInv(x_s,lambda_1)/(lambda_1*exponentialInv(x_s,1)):
            x = exponentialInv(x_s,lambda_1)
    return x
```

### Distribuci贸n de Poisson por el M茅todo de Transformada Inversa

**Funci贸n para generar una variable perteneciente a la distribuci贸n de Poisson**

El algoritmo de la funci贸n est谩 basado en el algoritmo para la generaci贸n de variables de Poisson por el m茅todo de transformada inversa descrito en: http://www.columbia.edu/~ks20/4404-Sigman/4404-Notes-ITM.pdf

```python
def poissonSim():
    x = 0
    p = np.exp(-lambda_0)
    s = p
    u = random.uniform(0,1)
    while u>s:
        x += 1
        p = p*(lambda_0/x)
        s += p
    return x
```

## Definici贸n de objetos y funciones para la simulaci贸n

### Modelo de objetos y m茅todos

Clase para la generaci贸n de clientes. Esta generaci贸n sigue la distribuci贸n de Poisson con par谩metro $\lambda$ (`lambda_0`).

```python
class Caja(object):
    def __init__(self, env, number): # tiempo_atender: expSim
        self.env = env
        self.numero = number
        self.cajero = simpy.Resource(env, capacity=1)
        self.tiempo_atender = 0
        self.tiempos_atender_clientes = []
        self.tiempos_espera = []
        self.cola = 0
        self.clientes_atendidos = 0
    
    def atender(self, cliente, tiempo_espera):
        yield self.env.timeout(self.tiempo_atender)
        self.clientes_atendidos += 1
        self.cola -= 1
        self.tiempos_espera.append(tiempo_espera)
        
    def aumentarCola(self):
        self.cola += 1
    
    def setTiempoAtender(self, tiempo_atender):
        self.tiempo_atender = tiempo_atender
        self.tiempos_atender_clientes.append(tiempo_atender)

def cliente(env, nombre, tiempo_atender, cajas):
    shortest = cajas[0]
    for caja in cajas:
        print("\U0001F4B2 Caja %d \U0001F4B2 %s" % (caja.numero, "" if caja.cola==0 else "\U0001F6B6"*caja.cola))
        if caja.cola < shortest.cola:
            shortest = caja
    with shortest.cajero.request() as req:
        arrive = env.now
        print("%d: %s hace cola en la caja %d \U0001F551" % (arrive, nombre, shortest.numero))
        shortest.aumentarCola()
        yield req
        shortest.setTiempoAtender(tiempo_atender)
        wait_time = env.now - arrive
        print("%d: %s llega al cajero (caja %d) \U0001F4B3" % (env.now, nombre, shortest.numero))
        yield env.process(shortest.atender(nombre, wait_time)) # esperamos a que nos atienda a nosotros
        print("%d: %s sale de la caja %d \U0001F9FE" % (env.now, nombre, shortest.numero))

envi = simpy.Environment()

cajas = [Caja(envi, i) for i in range(1,M+1)]

def setup(env): # TODO: 3 remover tiempo_atender y tiempo_llegada_clientes
    clientes = 0
    while True:
        clientes += 1
        yield env.timeout(1) # TODO: 1 sustituir "1" con poissonSim
        print("%d: Cliente %d llega a la tienda \U0001F6D2" % (env.now, clientes))
        env.process(cliente(envi, "Cliente %d" % clientes, clientes, cajas)) # TODO: 2 sustituir "clientes" con expSim
envi.process(setup(envi)) 
envi.run(until=20)
```

## Resultados de simulaci贸n

```python
all_tiempos_espera = []
all_tiempos_atender = []
all_clientes_atendidos = []

for caja in cajas:
    all_tiempos_espera.append(caja.tiempos_espera)
    all_tiempos_atender.append(caja.tiempos_atender_clientes)
    #for i in range(len(caja.tiempos_espera)):
        #all_tiempos_espera.append(caja.tiempos_espera[i])
    #for i in range(len(caja.tiempos_atender_clientes)):
        #all_tiempos_atender.append(caja.tiempos_atender_clientes[i])
    all_clientes_atendidos.append(caja.clientes_atendidos)

for caja in cajas:
    print("\U0001F4B2 CAJA %d \U0001F4B2" % caja.numero)
    print("   \U0001F551 Tiempos de espera en la cola: %s, promedio de tiempo de espera en cola %f" % (caja.tiempos_espera,np.mean(caja.tiempos_espera)))
    print("   \U0001F4B3 Tiempos de atender del cajero: %s, promedio que tarda el cajero en atender %f" % (caja.tiempos_atender_clientes,np.mean(caja.tiempos_atender_clientes)))
    print("   \U0001F9FE Total de clientes atendidos: %d, grado de utilizaci贸n de la caja: %f%%" % (caja.clientes_atendidos, caja.clientes_atendidos*100/np.sum(all_clientes_atendidos)))
    print("   \U0001F6B6 N煤mero de clientes en la cola:", caja.cola)
```
